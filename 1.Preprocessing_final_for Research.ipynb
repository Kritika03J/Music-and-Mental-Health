```python
data['While working'].fillna('Yes', inplace = True)
data['Instrumentalist'].fillna('No', inplace = True)
data['Primary streaming service'].fillna('Spotify', inplace = True)
data['Foreign languages'].fillna('Yes', inplace = True)
```

```python
data.isnull().sum()
```

```python
# Finding median of Age column to fill its missing values
m_age = data['Age'].median()
data['Age'].fillna(m_age, inplace = True)
```

```python
data.isnull().sum()
```

```python
# Finding number of missing bpm records with respect to favorite genre
null_bpm_counts = data[data['BPM'].isnull()].groupby('Fav genre').size().reset_index(name = 'null_bpm_count')
print(null_bpm_counts)
```

```python
# Finding mean of bpm with favorite genre grouping
mean_bpm_by_genre = data.groupby('Fav genre')['BPM'].mean().reset_index().round(2)
mean_bpm_by_genre.columns = ['Fav_genre', 'mean_bpm']
print(mean_bpm_by_genre)
```

```python
# We can see that Video Game Music has an outlier causing the mean value to get skewed, so we will find it and then remove it.
data_vgm = data[data['Fav genre'] == 'Video game music']

# Create a boxplot to check how many outliers are present in Video Game Music related bpm
plt.figure(figsize = (8, 6))
sns.boxplot(y = data_vgm['BPM'])
plt.title('Boxplot of BPM for Video Game Music Genre')
plt.ylabel('BPM')
plt.show()
```

There is only 1 outlier present in Video Game Music column.

```python
# Calculate the IQR for BPM values of genre Video Game Music
Q1 = data_vgm['BPM'].quantile(0.25)
Q3 = data_vgm['BPM'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Identify non-outlier BPM values for VGM
filtered_data_vgm = data_vgm[(data_vgm['BPM'].isna()) | ((data_vgm['BPM'] >= lower_bound) & (data_vgm['BPM'] <= upper_bound))]
non_vgm_data = data[data['Fav genre'] != 'Video game music']

# Concatenate non-outlier VGM records and non-VGM records
data = pd.concat([non_vgm_data, filtered_data_vgm])

# Resetting the index
data.reset_index(drop = True, inplace = True)
```

```python
# Finding mean of bpm with favorite genre grouping to check if vgm genre outlier is removed
mean_bpm_by_genre = data.groupby('Fav genre')['BPM'].mean().reset_index().round(2)
mean_bpm_by_genre.columns = ['Fav_genre', 'mean_bpm']
print(mean_bpm_by_genre)
```

```python
# Replace rows where BPM is NaN with its mean(i.e., those that were outliers and not included in filtered_data_vgm)
data['BPM'] = data.groupby('Fav genre')['BPM'].transform(lambda x: x.fillna(x.mean()))

# Resetting the index
data.reset_index(drop = True, inplace = True)
```

```python
vgm_data = data[data['Fav genre'] == 'Video game music'][['Fav genre', 'BPM']]
vgm_data['BPM'] = vgm_data['BPM'].round().astype(int)
```

```python
# Create a boxplot to check if outlier removed in Video Game Music related bpm
data_vgm = data[data['Fav genre'] == 'Video game music']
plt.figure(figsize = (8, 6))
sns.boxplot(y = data_vgm['BPM'])
plt.title('Boxplot of BPM for Video Game Music Genre')
plt.ylabel('BPM')
plt.show()
```

```python
# We can see that EDM has an outlier, so we will find it and then remove it.
data_edm = data[data['Fav genre'] == 'EDM']
Q1 = data_edm['BPM'].quantile(0.25)
Q3 = data_edm['BPM'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Identify non-outlier BPM values for EDM
filtered_data_edm = data_edm[(data_edm['BPM'].isna()) | ((data_edm['BPM'] >= lower_bound) & (data_edm['BPM'] <= upper_bound))]
non_edm_data = data[data['Fav genre'] != 'EDM']

# Concatenate non-outlier EDM records and non-EDM records
data = pd.concat([non_edm_data, filtered_data_edm])

# Resetting the index
data.reset_index(drop = True, inplace = True)
```

```python
# Finding outliers in BPM in individual genre
plt.figure(figsize = (10, 6))
sns.scatterplot(data = data, x = 'Fav genre', y = 'BPM', hue = 'Fav genre', palette = 'viridis', s = 100)
plt.title('Scatterplot of BPM with Different Colors for Each Genre')
plt.xlabel('Genre')
plt.ylabel('BPM')
plt.legend(title = 'Genre')
plt.show()
```

```python
# Converting the timestamp column to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'], format = 'mixed', errors = 'coerce')
print(data['Timestamp'])

# Sorting the DataFrame by the timestamp column
data = data.sort_values(by = 'Timestamp')

# Resetting the index
data = data.reset_index(drop = True)
```

```python
# Removing time stamp, permission and Primary streaming service columns
columns_to_remove = ['Timestamp', 'Permissions', 'Primary streaming service']
data = data.drop(columns = columns_to_remove)
```
